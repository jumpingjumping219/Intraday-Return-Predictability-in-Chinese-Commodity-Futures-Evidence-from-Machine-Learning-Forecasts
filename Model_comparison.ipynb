```python
#### import numpy as np
import pandas as pd
from scipy.stats import t
from itertools import combinations

# ---------------------------------------------------------------------
# 1. Diebold–Mariano test (two-sided, loss = |e|^power   or   e**2 )
# ---------------------------------------------------------------------
def dm_test(y_true, y_hat_1, y_hat_2, h=1, power=2):
    """
    Return (DM-stat, p-value) comparing model-1 vs model-2
    following Diebold & Mariano (1995).

    h     : forecast horizon
    power : 1 → absolute error, 2 → squared error (RMSE)
    """
    # 1) loss differential d_t
    e1, e2 = y_true - y_hat_1, y_true - y_hat_2
    if power == 1:
        d = np.abs(e1) - np.abs(e2)
    else:
        d = e1**2 - e2**2

    T = len(d)
    d_bar = d.mean()

    # 2) Newey–West variance with truncation lag h-1
    def autocov(series, k):
        return np.cov(series[k:], series[:-k])[0, 1] if k else series.var(ddof=1)

    gamma = np.array([autocov(d, k) for k in range(h)])
    var_d = gamma[0] + 2 * gamma[1:].sum()

    dm_stat = d_bar / np.sqrt(var_d / T)

    # 3) Two-sided p-value under t_{T-1}
    p_val = 2 * t.sf(np.abs(dm_stat), df=T-1)
    return dm_stat, p_val

# ---------------------------------------------------------------------
# 2. Build the DM matrix with significance stars
# ---------------------------------------------------------------------
def dm_table(y_true, pred_dict, h=1, power=2, stars=True):
    """
    Return a DataFrame whose (row, col) entry is the DM t-statistic
    comparing *row-model* vs *col-model* (positive ⇒ row wins).
    """
    models = list(pred_dict.keys())
    K = len(models)
    mat = np.zeros((K, K))

    for i, j in combinations(range(K), 2):
        dm_stat, p_val = dm_test(y_true,
                                 pred_dict[models[i]],
                                 pred_dict[models[j]],
                                 h=h, power=power)
        # row i vs col j
        mat[i, j] = dm_stat
        mat[j, i] = -dm_stat     # antisymmetric

    tab = pd.DataFrame(mat, index=models, columns=models)

    if stars:
        def starify(val):
            if val == 0:                       # diagonal
                return ""
            p = 2 * t.sf(abs(val), df=len(y_true)-1)
            if p < 0.01:  return f"{val:.2f}***"
            if p < 0.05:  return f"{val:.2f}**"
            if p < 0.10:  return f"{val:.2f}*"
            return f"{val:.2f}"
        tab = tab.applymap(starify)

    return tab
```

### following hypothesis test for each commodity futures on the out-of-sample data 

```python
import pandas as pd
pred_rf = pd.read_parquet('/mnt/rf_all_1min.parquet', engine = 'fastparquet')
pred_las = pd.read_csv('all_rolling_predictions_lasso_1min.csv')

codes = ['A', 'AG', 'AL', 'AP', 'AU', 'B', 'BU', 'C', 'CF', 'CS', 'CU', 'CY',
       'FB', 'FG', 'HC', 'J', 'JD', 'L', 'M', 'MA', 'NI', 'OI', 'PB', 'PP',
       'RB', 'RM', 'RU', 'SF', 'SM', 'SN', 'SR', 'TA', 'V', 'Y', 'ZC', 'ZN']
pred_rf['code'] = (
    pred_rf.groupby('datetime').cumcount()          
      .map(dict(enumerate(codes)))
)

pred_las['code'] = (
    pred_las.groupby('datetime').cumcount()          
      .map(dict(enumerate(codes)))
)
```

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from scipy import stats

def plot_code_comparison(df_las: pd.DataFrame, df_other: pd.DataFrame,
                         title_suffix="Other Model"):
    """
    绘制 LASSO 与其它模型 R²_OS 差异 + 显著性星号
    df_las  : 必含 ['datetime','y_true','y_pred','code']
    df_other: 同上
    """
    for _df in (df_las, df_other):
        _df["datetime"] = pd.to_datetime(_df["datetime"])

    def r2_os(df):
        sse_m = ((df.y_true - df.y_pred)**2).sum()
        sse_0 = (df.y_true**2).sum()
        return 1 - sse_m / sse_0 if sse_0 > 1e-12 else 0.0

    def one_sided_p(diff):
        t, p2 = stats.ttest_1samp(diff, 0.0, nan_policy="omit")
        return p2 / 2 if t > 0 else 1.0

    records = []
    for code in sorted(set(df_las.code) & set(df_other.code)):
        las = df_las[df_las.code == code].sort_values("datetime")
        oth = df_other[df_other.code == code].sort_values("datetime")
        merged = las[["datetime","y_true","y_pred"]].merge(
                     oth[["datetime","y_pred"]],
                     on="datetime", suffixes=("_las","_oth"))
        if len(merged) < 10: continue
        diff_r2 = r2_os(merged[["y_true","y_pred_las"]].rename(columns={"y_pred_las":"y_pred"})) - \
                  r2_os(merged[["y_true","y_pred_oth"]].rename(columns={"y_pred_oth":"y_pred"}))
        p_val  = one_sided_p((merged.y_true-merged.y_pred_oth)**2 -
                             (merged.y_true-merged.y_pred_las)**2)
        records.append({"code":code, "difference":diff_r2, "p":p_val})

    diff_df = pd.DataFrame(records).sort_values("difference").reset_index(drop=True)

    # ---------------- 绘图 ----------------
    plt.style.use("default")  
    fig, ax = plt.subplots(figsize=(14,6))

    bars = ax.bar(diff_df["code"], diff_df["difference"],
                  color="#7f7f7f", edgecolor="black", linewidth=0.8)

    for i, row in diff_df.iterrows():
        star, color = "", None
        if row.p < 0.01: star, color = "***", "#ff7f0e"
        elif row.p < 0.05: star, color = "**", "#2ca02c"
        elif row.p < 0.10: star, color = "*",  "#1f77b4"
        if star:
            y_pos = row.difference + (0.003 if row.difference>=0 else -0.003)
            ax.text(i, y_pos, star, ha="center",
                    va="bottom" if row.difference>=0 else "top",
                    color=color, fontsize=13, fontweight="bold")

    ax.axhline(0, color="black", linewidth=1)
    ax.set_ylabel(r"$R^2_{OS}(\mathrm{LASSO}) - R^2_{OS}(\mathrm{Other})$", fontsize=13)
    ax.set_xlabel("Commodity Code", fontsize=13)
    ax.set_title(fr"LASSO vs {title_suffix}:  $R^2_{{OS}}$ Difference by Commodity",
                 fontsize=15, pad=15)
    ax.set_xticklabels(diff_df["code"], rotation=45, ha="right")
    ax.set_ylim(min(-0.15, diff_df["difference"].min()-0.02),
                diff_df["difference"].max()+0.03)
    ax.grid(False)       
    plt.tight_layout()
    plt.show()

    return diff_df

result_df = plot_code_comparison(pred_las, pred_rf, title_suffix="RandomForest")
print(result_df.head())
```
